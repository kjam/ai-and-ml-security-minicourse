### Introduction to OpenGuardrails

I wanted to highlight some guardrails projects that aren't built by huge AI vendors to get you thinking more about how to both use systems that have open licenses and train your own. In doing so, I came across OpenGuardrails, which is built off of research from a small team who wanted to build a different approach to modeling guardrails. 


- Models at: [https://huggingface.co/openguardrails](https://huggingface.co/openguardrails)
- GitHub for [openguardrails](https://github.com/openguardrails)
- Original [Research Paper](https://arxiv.org/abs/2510.19169)


```python
from openai import OpenAI
```

```python
openai_api_key = "EMPTY"
openai_api_base = "http://192.168.1.176:8000/v1" # change this for your setup!
```

```python
client = OpenAI(
    api_key=openai_api_key,
    base_url=openai_api_base,
)
```

```python
def call_openguardrails(client, prompt):
    completion = client.chat.completions.create(
        model="openguardrails/OpenGuardrails-Text-2510",
        messages=[{
            "role": "user",
            "content": prompt
        }]
    )
    return completion.choices[0].message.content
    
```

```python
CATEGORIES = {
    'S1': 'General political',
    'S2': 'Sensitive political', 
    'S3': 'Insult to national symbols or leaders', 
    'S4': 'Harm to minors', 
    'S5': 'Violent crime', 
    'S6': 'Non-violent crime',
    'S7': 'Pornography', 
    'S8': 'Hate and discrimination',
    'S9': 'Prompt attacks',
    'S10': 'Profanity',
    'S11': 'Privacy invasion',
    'S12': 'Commercial violation',
    'S13': 'IP Infringement',
    'S14': 'Harrassment',
    'S15': 'Weapons of mass destruction',
    'S16': 'Self harm',
    'S17': 'Sexual crimes',
    'S18': 'Threats',
    'S19': 'Professional advice',
}
```

```python
def parse_category(response):
    category = response.split('\n')[1]
    return CATEGORIES[category]
```

```python
def call_guardrail_and_parse(client, prompt):
    response = call_openguardrails(client, prompt)
    if 'S' in response:
        return "Unsafe: {}".format(parse_category(response))
    return response
```
